IMDB Data scraping

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.imdb.com/list/ls081113000/'
response = requests.get(url)
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
print(soup)
```

```python
movie_name = []
year = []
duration = []
genre = []
ratings = []
description =[]
# director = []
# stars = []
```

```python
movie_data = soup.findAll('div', attrs={'class': 'lister-item mode-detail'})
movie_data
```

```python
for store in movie_data:
  name = store.h3.a.text
  movie_name.append(name)

  release_year = store.h3.find('span', class_ = 'lister-item-year text-muted unbold').text.replace('(', '').replace(')', '')
  year.append(release_year)

  time_duration = store.p.find('span', class_ = 'runtime').text
  duration.append(time_duration)

  movie_genre = store.p.find('span', class_ = 'genre').text.replace('\n', '').replace(' ', '')
  genre.append(movie_genre)

  movie_ratings = store.find('div', class_= 'ipl-rating-star small').text.replace('\n\n\n\n\n\n\n\n', '').replace('\n', '')
  ratings.append(movie_ratings)
  para = store.find('p', class_ = '').text.replace('\n', '')
  description.append(para)
```

```python
movie_DF = pd.DataFrame({'Name of movie': movie_name, 'Year of relase': year, 'Watchtime': duration, 'Movie genre': genre, 'Movie Rating': ratings, 'Description': description})
movie_DF
```
